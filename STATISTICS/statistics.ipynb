# Column Category Classification – loan_approved.csv

Loan_ID  
**Category Type:** Nominal  
**Reason:** This is only an identification code used to uniquely identify each record. It does not carry any numerical meaning or ranking.

Gender  
**Category Type:** Nominal  
**Reason:** This column contains categories such as *Male* and *Female*. There is no natural order between these categories.

Married  
**Category Type:** Nominal  
**Reason:** Values such as *Yes* and *No* are just labels and do not have any ranking or order.

Dependents  
**Category Type:** Discrete  
**Reason:** This column represents a count of dependents (0, 1, 2, 3…). These are whole numbers and cannot be in fractions.

Education  
**Category Type:** Nominal  
**Reason:** Categories like *Graduate* and *Not Graduate* are just names and do not follow any ranking.

Self_Employed  
**Category Type:** Nominal  
**Reason:** Indicates whether a person is self-employed (Yes/No). There is no ranking involved.

ApplicantIncome  
**Category Type:** Numerical (Continuous)  
**Reason:** Represents income of the applicant. This value can take a wide range of numbers and is treated as continuous data.

CoapplicantIncome  
**Category Type:** Numerical (Continuous)  
**Reason:** Income of the co-applicant. This is also a continuous numeric value.

LoanAmount  
**Category Type:** Numerical (Continuous)  
**Reason:** The loan amount is measured numerically and can take many possible values.

Loan_Amount_Term  
**Category Type:** Discrete  
**Reason:** This column contains fixed term values such as 120, 180, 240, and 360 months. These are countable and come from a limited set.

Credit_History  
**Category Type:** Nominal  
**Reason:** This usually contains 0 or 1 (No / Yes). Even though it is numeric in form, it represents categories, not an ordered ranking.

Property_Area  
**Category Type:** Nominal  
**Reason:** Categories such as *Urban*, *Semiurban*, and *Rural*. There is no inherent order between them.

Loan_Status  
**Category Type:** Nominal  
**Reason:** This is the output/target column indicating whether the loan is approved (Y) or not approved (N).

```python

# -------------------------------
# NUMERICAL COLUMNS
# -------------------------------

# ApplicantIncome → Median
df["ApplicantIncome"] = df["ApplicantIncome"].fillna(df["ApplicantIncome"].median())

# CoapplicantIncome → Mean
df["CoapplicantIncome"] = df["CoapplicantIncome"].fillna(df["CoapplicantIncome"].mean())

# LoanAmount → Median
df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].median())

# Loan_Amount_Term → Mode
df["Loan_Amount_Term"] = df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].mode()[0])

# Credit_History → Mode
df["Credit_History"] = df["Credit_History"].fillna(df["Credit_History"].mode()[0])


# -------------------------------
# CATEGORICAL COLUMNS
# -------------------------------

df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
df["Married"] = df["Married"].fillna(df["Married"].mode()[0])
df["Dependents"] = df["Dependents"].fillna(df["Dependents"].mode()[0])
df["Education"] = df["Education"].fillna(df["Education"].mode()[0])
df["Self_Employed"] = df["Self_Employed"].fillna(df["Self_Employed"].mode()[0])
df["Property_Area"] = df["Property_Area"].fillna(df["Property_Area"].mode()[0])

# Correct target column name
df["Loan_Status (Approved)"] = df["Loan_Status (Approved)"].fillna(
    df["Loan_Status (Approved)"].mode()[0]
)
```
